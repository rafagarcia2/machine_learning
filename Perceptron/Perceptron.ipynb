# Exercício sobre Perceptron
Ajustando os dados de uma clusterização. Percebendo padrões e agrupando novos dados.

```
# Importando as libraries
import pandas as pd
import tensorflow as tf

import numpy as np
```

```
# declarando as variáveis
w = [0.4, -0.6, 0.6] # pesos
limiar = 0.5

tx_aprend = 0.4 # taxa de aprendizado
d = -1 # valor esperado
```

## Testando a manipulação de vetores com um exemplo

```
x1 = [0, 0, 1]  # Exemplo 001

m = np.multiply(x1,w)
u = np.sum(m) - limiar

print(u)
```

## Criando função para treino

```
def clustering(w, vetor):
  m = np.multiply(w, vetor)
  u = np.sum(m) - limiar
  
  # verificação
  y = 1 if (u >= 0) else -1
  
  for i in range(len(w)):
    w[i] = w[i] + tx_aprend * vetor[i] * (d - y)
  
  new_limiar = limiar + tx_aprend * -1 * (d - y)
  
  print("Limiar: ", new_limiar)
  print("Pesos", w)
  print("Resultado (y): ", y)
  
  return (new_limiar, y, w)
```

## Treinando os dados
Agora vamos rodar os dados para treinar nossa rede neural.

```
limiar, d, w = clustering(w, [0, 0, 1]) # para 001
```

```
limiar, d, w = clustering(w, [1, 1, 0])  # para 110
```

```
# Segunda Rodada de treinamento
limiar, d, w = clustering(w, [0, 0, 1]) # para 001
print("\nSEGUNDO CASO\n")
limiar, d, w = clustering(w, [1, 1, 0])  # para 110
```

## Testando nosso modelo
Agora vamos testar nosso modelo com novos dados

```
print("Saída de 111: ")
limiar, d, w = clustering(w, [1, 1, 1])
print("Saída de 000: ")
limiar, d, w = clustering(w, [0, 0, 0])
print("Saída de 100: ")
limiar, d, w = clustering(w, [1, 0, 0])
print("Saída de 011: ")
limiar, d, w = clustering(w, [0, 1, 1])
```
